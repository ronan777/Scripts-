```python
from random import random
from random import seed
import matplotlib.pyplot as plt

def AlicePoint(n1,n2):

  if n1 == "H" and n2 == "H":
      return 1
  else:
      return 0

def BobPoint(n1,n2):

  if n1 == "H" and n2 == "T":
      return 1
  else:
      return 0


def CreateCoinTossSequence(length):
    sequence = [];
    seed();
    for i in range(length):
        toss = random()
        if(toss >= 0.5):
            sequence.append("H");
        else:
            sequence.append("T");
    return sequence;

def RunGame(length):
    alicePoints = 0;
    bobPoints = 0;
    sequence = CreateCoinTossSequence(length);
#    print(sequence);
    for i in range(length-1):
        alicePoints = alicePoints + AlicePoint(sequence[i],sequence[i+1]);
        bobPoints = bobPoints + BobPoint(sequence[i], sequence[i+1]);

    return alicePoints,bobPoints;


Xdata = [];
Y1data = [];
Y2data = [];
avalice=0.;
avbob=0.;
aliceWins = 0.;
bobWins = 0.;
tie=0.;
#GAME_LENGTH = 4;
NUMBER_OF_GAMES = 2000;
MAX_LENGTH = 500

for j in range(MAX_LENGTH):

#    NUMBER_OF_GAMES = j+1;
    GAME_LENGTH = 2+j;
    aliceWins = 0.;
    bobWins = 0.
    tie = 0.;
    for i in range(NUMBER_OF_GAMES):
      alicePoints,bobPoints = RunGame(GAME_LENGTH)
#        print (alicePoints,bobPoints);
      if(alicePoints > bobPoints):
        aliceWins += 1;
      elif (alicePoints < bobPoints):
        bobWins += 1;
      else :
        tie += 1;

    bobWinPercent = bobWins/NUMBER_OF_GAMES*100;
    aliceWinPercent  = aliceWins / NUMBER_OF_GAMES * 100;
    tiePercent = tie/ NUMBER_OF_GAMES*100;
    print('Alice wins: % 6.2f' % aliceWinPercent, "%")
    print('Bob wins: % 6.2f' % bobWinPercent, "%")
    print('Tie: % 6.2f' % tiePercent,"%")
    print(j)
    Xdata.append(j);
    Y1data.append(aliceWinPercent);
    Y2data.append(bobWinPercent);
    avalice+=aliceWinPercent;
    avbob+=bobWinPercent;
avalice/= MAX_LENGTH;
avbob/= MAX_LENGTH;
print('averages: % 6.2f Alice' % avalice, '% 6.2f Bob' % avbob);
plt.plot(Xdata,Y1data,Xdata,Y2data);
plt.legend(['Alice wins %',"Bob wins %"]);
plt.show()
```
